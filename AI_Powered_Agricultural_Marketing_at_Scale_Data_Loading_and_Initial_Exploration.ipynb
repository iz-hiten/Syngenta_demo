### Re-running: Threshold Optimization using Precision-Recall Curve

```python
from sklearn.metrics import precision_recall_curve, auc
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix

# Get probabilities for the positive class (clicked_status=True)
y_probs = best_xgb_model.predict_proba(X_test)[:, 1]

# Calculate precision, recall, and thresholds
precision, recall, thresholds = precision_recall_curve(y_test, y_probs)

# Calculate AUC for Precision-Recall Curve
pr_auc = auc(recall, precision)

print(f'Precision-Recall AUC: {pr_auc:.2f}')

# Plot the Precision-Recall Curve
plt.figure(figsize=(10, 6))
plt.plot(recall, precision, label=f'Precision-Recall curve (AUC = {pr_auc:.2f})')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve for Tuned XGBoost Model')
plt.legend(loc='lower left')
plt.grid(True)
plt.show()

# Find the threshold that maximizes the F1-score
f1_scores = 2 * (precision * recall) / (precision + recall)
f1_scores = np.nan_to_num(f1_scores) # Handle potential division by zero
optimal_threshold_idx = np.argmax(f1_scores)
optimal_threshold = thresholds[optimal_threshold_idx]

print(f"Optimal Threshold (maximizing F1-score): {optimal_threshold:.4f}")
print(f"Precision at optimal threshold: {precision[optimal_threshold_idx]:.2f}")
print(f"Recall at optimal threshold: {recall[optimal_threshold_idx]:.2f}")

# Evaluate the model with the optimal threshold
y_pred_optimal_threshold = (y_probs >= optimal_threshold).astype(int)

print('\nClassification Report (Tuned XGBoost with Optimal Threshold):')
print(classification_report(y_test, y_pred_optimal_threshold))

print('\nConfusion Matrix (Tuned XGBoost with Optimal Threshold):')
print(confusion_matrix(y_test, y_pred_optimal_threshold))
```

### Re-running: Generating Personalized Marketing Recommendations

```python
# First, let's get the probabilities and predictions for the entire merged_campaign_grower_df
# We need to ensure X_processed is aligned with merged_campaign_grower_df

# Re-create X using the original merged_campaign_grower_df to get all features
X_full = merged_campaign_grower_df.drop([
    'id', 'grower_id', 'message_sent_date', 'grower_crop_calendar',
    'product_scan_datetime', 'campaign_attendance_date', 'parsed_crop_calendar',
    'delivered_status', 'opened_status', 'clicked_status',
    'delivery_rate', 'open_rate', 'click_rate', # These are per-campaign engagement rates and directly reveal target
    'overall_delivery_rate', 'overall_open_rate', 'overall_click_rate' # These are overall grower engagement rates and also leak information
], axis=1)

# Apply the same preprocessing steps to X_full as applied to X_train/X_test
X_full_processed = X_full.select_dtypes(include=['number', 'bool'])
X_full_processed = X_full_processed.drop(columns=['product_scan_month', 'campaign_attendance_month'], errors='ignore')

# Ensure the columns in X_full_processed match the columns the model was trained on (X_train_smote)
# This is important if some columns were dropped during feature selection or if order changed
missing_cols = set(X_train_smote.columns) - set(X_full_processed.columns)
for c in missing_cols:
    X_full_processed[c] = 0  # Add missing columns with default value (e.g., 0)
X_full_processed = X_full_processed[X_train_smote.columns] # Ensure the order of feature columns is the same


y_full_probs = best_xgb_model.predict_proba(X_full_processed)[:, 1]
y_full_pred_optimal = (y_full_probs >= optimal_threshold).astype(int)

# Add predictions and probabilities back to the main DataFrame
merged_campaign_grower_df['predicted_click_proba'] = y_full_probs
merged_campaign_grower_df['predicted_click'] = y_full_pred_optimal

# Generate recommendations for each grower
def generate_recommendations(grower_data):
    grower_id = grower_data.name # Access grower_id from the group name
    state = grower_data['state'].iloc[0]
    language = grower_data['language'].iloc[0]
    crop = grower_data['crop_calendar_crop'].iloc[0]
    season = grower_data['crop_calendar_season'].iloc[0]

    # Sort by predicted click probability in descending order
    grower_data = grower_data.sort_values(by='predicted_click_proba', ascending=False)

    # Get top recommended campaigns for this grower (based on high probability and predicted click)
    top_campaigns = grower_data[grower_data['predicted_click'] == 1].head(3)

    recommendations = []
    if not top_campaigns.empty:
        for idx, row in top_campaigns.iterrows():
            recommendations.append(f"Product: {row['campaign_product']} for {row['campaign_crop']} (Probability: {row['predicted_click_proba']:.2f})")
    else:
        recommendations.append("No specific campaigns recommended based on high click probability. Consider general awareness campaigns.")

    return {
        'grower_id': grower_id,
        'state': state,
        'language': language,
        'crop': crop,
        'season': season,
        'recommended_campaigns': recommendations
    }

# Group by grower and apply the recommendation function
grower_recommendations = merged_campaign_grower_df.groupby('grower_id').apply(generate_recommendations, include_groups=False).reset_index(drop=True)

# Convert the Series of dictionaries to a DataFrame
grower_recommendations_df = pd.DataFrame(grower_recommendations.tolist())

print("Generated personalized marketing recommendations for growers.")
display(grower_recommendations_df.head())


print("\nExample recommendations for a sample grower (e.g., GRW_00001):")
sample_grower_id = 'GRW_00001'
display(grower_recommendations_df[grower_recommendations_df['grower_id'] == sample_grower_id].iloc[0]['recommended_campaigns'])
```

```python
grower_engagement_summary = merged_campaign_grower_df.groupby('grower_id').agg(
    total_messages_sent=('id', 'count'),
    total_messages_delivered=('delivered_status', lambda x: x.sum()),
    total_messages_opened=('opened_status', lambda x: x.sum()),
    total_messages_clicked=('clicked_status', lambda x: x.sum())
).reset_index()

grower_engagement_summary['overall_delivery_rate'] = grower_engagement_summary['total_messages_delivered'] / grower_engagement_summary['total_messages_sent']
grower_engagement_summary['overall_open_rate'] = grower_engagement_summary['total_messages_opened'] / grower_engagement_summary['total_messages_delivered']
grower_engagement_summary['overall_click_rate'] = grower_engagement_summary['total_messages_clicked'] / grower_engagement_summary['total_messages_opened']

# Handle potential division by zero for rates
grower_engagement_summary.fillna(0, inplace=True)

print('Grower engagement history features created.')
display(grower_engagement_summary.head())
```

### 2. Parse `grower_crop_calendar` in `growers_df`

```python
import json

def parse_crop_calendar(calendar_str):
    try:
        return json.loads(calendar_str)
    except (json.JSONDecodeError, TypeError):
        return {}

growers_df['parsed_crop_calendar'] = growers_df['grower_crop_calendar'].apply(parse_crop_calendar)

# Extracting common fields
growers_df['crop_calendar_season'] = growers_df['parsed_crop_calendar'].apply(lambda x: x.get('season'))
growers_df['crop_calendar_crop'] = growers_df['parsed_crop_calendar'].apply(lambda x: x.get('crop'))
growers_df['crop_calendar_sowing_date'] = growers_df['parsed_crop_calendar'].apply(lambda x: x.get('sowing_date'))
growers_df['crop_calendar_harvest_date'] = growers_df['parsed_crop_calendar'].apply(lambda x: x.get('harvest_date'))

# Convert date strings within the parsed calendar to datetime objects as well
growers_df['crop_calendar_sowing_date'] = pd.to_datetime(growers_df['crop_calendar_sowing_date'], errors='coerce')
growers_df['crop_calendar_harvest_date'] = pd.to_datetime(growers_df['crop_calendar_harvest_date'], errors='coerce')

print('Parsed `grower_crop_calendar` and extracted season, crop, sowing_date, harvest_date.')
print('\nGrowers_df head with new crop calendar columns:')
display(growers_df[['grower_id', 'grower_crop_calendar', 'crop_calendar_season', 'crop_calendar_crop', 'crop_calendar_sowing_date', 'crop_calendar_harvest_date']].head())
print('\nGrowers_df info after parsing crop calendar:')
print(growers_df.info())
```

1. Data Loading and Initial Exploration
We started by loading four essential datasets into pandas DataFrames: whatsapp_campaign.csv, digital_funnel_weekly.csv, growers.csv, and retailers.csv. For each dataset, we displayed its shape (number of rows and columns), the first few rows using .head(), and a concise summary of its structure and data types using .info(). This initial step helped us understand the basic characteristics of each dataset.

2. Data Preprocessing
Date Column Conversion: We converted relevant columns from object (string) type to datetime objects (message_sent_date, week_start_date, product_scan_datetime, campaign_attendance_date). This is crucial for performing time-series analysis and extracting temporal features.
Parsing grower_crop_calendar: The grower_crop_calendar column in growers_df contained JSON-like strings. We wrote a function to parse these strings, extracting season, crop, sowing_date, and harvest_date into separate columns. The extracted dates were also converted to datetime objects.
Handling Missing Values in growers_df: We addressed missing values in several columns within growers_df. Categorical columns like crop_calendar_season, crop_calendar_crop, and product_name had their NaN values filled with 'Unknown' or 'No Product Scanned'. For grower_farm_size, missing values were imputed with the median farm size to maintain data integrity without introducing significant bias.
3. Data Integration
We merged whatsapp_campaign_df with growers_df using grower_id as the key. This step enriched the campaign data with detailed grower demographic and historical information, creating a comprehensive merged_campaign_grower_df that serves as the foundation for our modeling.

4. Exploratory Data Analysis (EDA)
We conducted EDA to gain insights into our data:

WhatsApp Campaign Engagement: We visualized the distribution of message delivery, opened, and clicked statuses. We also analyzed the average open rates by campaign_crop and average click rates by campaign_product to identify engagement patterns.
Digital Funnel Performance: We explored trends in social_post_impression and lead_form_submission over time, and examined the average lead conversion rate by campaign_crop for digital campaigns.
Grower Demographics: We analyzed the distribution of growers by state, grower_age, language, device_type, gender, product_scan, and offline_campaign_attended to understand our target audience's profile.
5. Feature Engineering
We created new features to enhance our model's predictive power:

Engagement Rates: Calculated delivery_rate, open_rate, and click_rate for the whatsapp_campaign_df.
Conversion Rates: Computed lp_visit_rate (landing page visit rate) and lead_conversion_rate for digital_funnel_weekly_df.
Temporal Features: Extracted dayofweek and month from various date columns (message_sent_date, week_start_date, product_scan_datetime, campaign_attendance_date).
Grower Engagement History Features: Summarized grower engagement by calculating total_messages_sent, total_messages_delivered, total_messages_opened, and total_messages_clicked for each grower. We then derived overall_delivery_rate, overall_open_rate, and overall_click_rate to capture long-term engagement behavior.
Age Bins and Farm Size Categories: Created categorical bins for grower_age (grower_age_bin) and grower_farm_size (farm_size_category) to capture non-linear relationships and potentially improve model performance.
Categorical Feature Encoding: Applied LabelEncoder to all relevant categorical columns (e.g., campaign_product, state, gender, crop_calendar_season) to convert them into numerical representations suitable for machine learning models.
6. Model Training and Evaluation
Defining Target Variables and Addressing Data Leakage: We defined clicked_status as our target variable (y). Crucially, we carefully selected features (X) by dropping columns that would directly or indirectly lead to data leakage (e.g., delivered_status, opened_status, per-campaign rates like delivery_rate, open_rate, click_rate, and overall engagement rates like overall_delivery_rate, overall_open_rate, overall_click_rate). This ensured the model learns from predictive features rather than target-related information.
Splitting Data: We split the preprocessed data into training and testing sets (X_train, X_test, y_train, y_test) using an 80/20 ratio and stratify=y to maintain the class distribution of the target variable in both sets.
Addressing Class Imbalance with SMOTE: Recognizing the significant class imbalance in our target variable (few clicks compared to non-clicks), we applied the Synthetic Minority Over-sampling Technique (SMOTE) to the training data (X_train_smote, y_train_smote) to create synthetic samples of the minority class, thus balancing the dataset for training.
Training Models: We trained and evaluated several classification models:
Logistic Regression: A baseline model was trained both on the original data and again on the SMOTE-balanced data. We evaluated its performance using classification reports, confusion matrices, and ROC AUC scores.
Random Forest: An ensemble model was trained on the SMOTE-balanced data, and its performance was similarly evaluated.
XGBoost: An advanced gradient boosting model was trained on the SMOTE-balanced data. Its performance was assessed using the same metrics.
Hyperparameter Tuning for XGBoost: We performed GridSearchCV to find the optimal hyperparameters for the XGBoost model, aiming to maximize the F1-score. This resulted in the best_xgb_model.
Threshold Optimization: We used the Precision-Recall curve to find an optimal_threshold that maximizes the F1-score for the best_xgb_model. This is important for imbalanced datasets, as the default 0.5 threshold might not be ideal.
7. Feature Importance and Model Interpretability
Feature Importance Analysis: For each trained model (Logistic Regression, Random Forest, XGBoost, and Tuned XGBoost), we analyzed and visualized the top 10 most important features. This helps us understand which factors contribute most to predicting a click.
SHAP Values for Tuned XGBoost: We used SHAP (SHapley Additive exPlanations) to interpret the best_xgb_model. We calculated SHAP values for the test set and visualized them using summary plots (bar and detailed) and a force plot for an individual prediction. SHAP provides a more nuanced understanding of how each feature influences model predictions, both globally and for individual instances.
8. Generating Personalized Marketing Recommendations
Finally, we integrated the retrained and tuned XGBoost model to generate personalized marketing recommendations. For each grower, we calculated the predicted click probability for all available campaigns. Campaigns with a predicted click (based on the optimal_threshold) were then sorted by their probability, and the top 3 were recommended. If no campaigns met the click prediction criteria, a general awareness recommendation was provided. The resulting grower_recommendations_df contains these personalized insights, along with an example recommendation for a sample grower.
